```
import pydot
from collections import defaultdict 
from IPython.display import Image, display
```

```
def lstr(p):
        return ''.join([str(elem)  for elem in p])
```

```
def ModifiedBFS(s): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                    node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append(d) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
def PNcalculate(graph, s):
  respectivePn = defaultdict(int)
  visited = defaultdict()
  s = sorted(s)
  queue = []
  queue.append(s)
  respectivePn[lstr(s)] = 0
  visited[lstr(s)] = True

  pn = 0
  while queue:
    s = queue.pop(0)    
    for d in graph[lstr(s)]:
      if (d != s) and not(lstr(d) in visited.keys()):
        queue.append(d)
        respectivePn[lstr(d)] = respectivePn[lstr(s)] + 1
        visited[lstr(d)] = True
        if(respectivePn[lstr(d)] > pn):
          print("Ahora el mayor es " + lstr(d))
          pn = respectivePn[lstr(d)]

  return pn
```

```
def NetworkPancake(s): 

       
        graph = defaultdict(list)
        visited = defaultdict()
        G = pydot.Dot(graph_type="graph")            
        queue = [] 
        queue.append(s) 
        visited[lstr(s)] = True
        
        while queue: 
          s = queue.pop(0)          
          if(s != sorted(s)):
              node = pydot.Node(lstr(s))
          else:
            node = pydot.Node(lstr(s),style="filled", fillcolor="green") # first sorted green
          G.add_node(node)
                    
          for i in range(len(s)-1):
            d = s[i+1::-1] + s[i+2:]  
            loMeto = lstr(s) not in graph[lstr(d)]

            if loMeto:
              edge = pydot.Edge(lstr(s),lstr(d))
              G.add_edge(edge)

            if (d != s) and not(lstr(d) in visited.keys()):
              
              if(d != sorted(d)):
                node = pydot.Node(lstr(d))
              else:
                node = pydot.Node(lstr(d),style="filled", fillcolor="green")
              
              G.add_node(node)  
              queue.append(d) 
              visited[lstr(d)] = True
            graph[lstr(s)].append(lstr(d))
                
        im = Image(G.create_png())
        display(im)
        return PNcalculate(graph, s)
```

```
def ModifiedBFSPreSols(s): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]

            flag = False
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                if(d == sorted(d)):
                     flag = True
            if flag:
              node = pydot.Node(lstr(s),style="filled", fillcolor="yellow")
            else: 
              node = pydot.Node(lstr(s))

            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                     node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append(d) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
def ModifiedBFSPreSolsLevel(s,L):  

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append([s,0]) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            e = queue.pop(0)
            s = e[0]
            l = e[1] 
            graph[lstr(s)]=[]

            flag = False
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                if(d == sorted(d)):
                     flag = True
            if flag:
              node = pydot.Node(lstr(s),style="filled", fillcolor="yellow")
            else: 
              node = pydot.Node(lstr(s))

            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)) and (l < L):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                     node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append([d,l+1]) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
def ModifiedBFSPreSols12h(s): 
       
        graph = defaultdict(list)
        visited = defaultdict() 
                
        queue = [] 
          
        queue.append(s) 
        visited[lstr(s)] = True
        
        while queue: 
 
          s = queue.pop(0) 
                   
          for i in range(len(s)-1):
            d = s[i+1::-1] + s[i+2:]  

            if (d != s) and not(lstr(d) in visited.keys()):
              queue.append(d) 
              visited[lstr(d)] = True
            graph[lstr(s)].append(lstr(d))                
                
        return PNcalculate(graph, s)
```

```
def NetworkPancakeDoubleSpatulas(s): 
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        queue = []  
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 

              s = queue.pop(0) 
              graph[lstr(s)]=[]
              flag = False
            
              for i in range(len(s)-1):
                  for j in range(i+1,len(s)):
                    d = s[:i]+s[j::-1][0:j-i+1] + s[j+1:]   
                    if(d == sorted(d)):
                          flag = True
                    if flag:
                      node = pydot.Node(lstr(s),style="filled", fillcolor="yellow")
                    else: 
                      node = pydot.Node(lstr(s))

              G.add_node(node)

              if(s != sorted(s)):
                for i in range(len(s)-1):
                  for j in range(i+1,len(s)):
                    d = s[:i]+s[j::-1][0:j-i+1] + s[j+1:]  
                    if (d != s) and not(lstr(d) in visited.keys()):
                  
                      if(d != sorted(d)):
                        node = pydot.Node(lstr(d))
                      else:
                        node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                      
                      G.add_node(node)
                      edge = pydot.Edge(lstr(s),lstr(d))
                      G.add_edge(edge)
                      graph[lstr(s)].append(lstr(d))
                      queue.append(d) 
                      visited[lstr(d)] = True 
                
        im = Image(G.create_png())
        display(im)
```
